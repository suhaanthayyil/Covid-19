# 01 — Exploratory Data Analysis

Load the COVID-19 dataset, keep confirmed positives, create the Long COVID target, and explore class distribution and feature correlations.

```python
import sys
import os
ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.insert(0, os.path.join(ROOT, 'src'))
os.chdir(ROOT)

from preprocess import load_raw_data, find_classification_column, create_target_and_filter
```

```python
# Load and filter to confirmed positive COVID cases; create LONG_COVID = 1 if PATIENT_TYPE == 2
raw = load_raw_data(data_dir=os.path.join(ROOT, 'data', 'raw'))
class_col = find_classification_column(raw)
df = create_target_and_filter(raw, class_col)
print(f"Filtered to {len(df):,} confirmed positive cases.")
df.head()
```

```python
# Class distribution: Long COVID (1) vs No (0)
import matplotlib.pyplot as plt
import seaborn as sns

counts = df['LONG_COVID'].value_counts().sort_index()
plt.figure(figsize=(6, 4))
plt.bar(counts.index.astype(str), counts.values, color=['#2ecc71', '#e74c3c'])
plt.xlabel('Long COVID')
plt.ylabel('Number of Patients')
plt.title('Long COVID vs. No Long COVID — Class Distribution')
plt.xticks([0, 1], ['0 (No)', '1 (Yes)'])
for i, v in enumerate(counts.values):
    plt.text(i, v + 0.01*counts.values.max(), str(v), ha='center')
plt.tight_layout()
plt.savefig(os.path.join(ROOT, 'figures', 'fig1_class_distribution.png'), dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Correlation heatmap (after preparing features — next notebook does full preprocessing)
from preprocess import prepare_features_and_target, FEATURE_COLS

feature_cols = [c for c in FEATURE_COLS if c in df.columns]
X, y = prepare_features_and_target(df, feature_cols=feature_cols, drop_pct_missing=0.30)
corr_df = X.copy()
corr_df['LONG_COVID'] = y.values
plt.figure(figsize=(10, 8))
sns.heatmap(corr_df.corr(method='pearson'), annot=True, fmt='.2f', cmap='RdBu_r', center=0, square=True)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig(os.path.join(ROOT, 'figures', 'fig2_correlation_heatmap.png'), dpi=300, bbox_inches='tight')
plt.show()
```
